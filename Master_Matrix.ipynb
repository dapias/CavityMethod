Here, I load the data stored for the Cavity Method using the master operator built up for a random regular graph with n = 1000 vertices and connectivity equal to c = 5, and compare it with the numerical diagonalization done for an ensemble of M = 1000 and T = 1.5.

```julia
cav1 = load("rho_master.jld")
```

```julia
rho_cav = cav1["rho"]
l_grid = cav1["lambdas"];
epsilon = cav1["epsilon"];
```

```julia
plot(l_grid, rho_cav,  ls = :dash, label = L"\epsilon"*"= 0.005im", xlabel = L"\lambda", ylabel = L"\rho(\lambda)")
```

```julia
numerics = load("emp_master.jld")
```

```julia
rho_emp = numerics["eigenvalues"];
```

```julia
stephist!(rho_emp, nbins = 100, normed = true, label = "Numerics", ylim = (0.,1,5))
```
