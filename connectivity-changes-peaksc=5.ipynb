```julia
Np = 2500
nsteps = 25000
epsilon = 1.e-300
ensemble = Int64(10^5);
```

```julia
epsilon2 = 1.e-3
```

## Overview

```julia
T = 0.1
lambda_barrat = -linspace(-0.,1.01,200)
```

```julia
c = 5
```

```julia
rho_ba = [rho_barrat_population(i, c, T, Np, ensemble, nsteps, epsilon, epsilon2) for i in lambda_barrat];
```

```julia
plot(lambda_barrat, rho_ba[:,1], label = "c = $(c)")
plt[:yscale]("log")
#plt[:xscale]("symlog")
plt[:grid](true)
plt[:gca]()[:xaxis][:grid](true, which="minor") 
```

## First peak

```julia
#lambda = -linspace(0.19,0.21,100)
lambda = logspace(-6,-1,101);
```

```julia
ep = [1.e-3, 1.e-4,1e-5,1.e-6]
```

### Left side

```julia
l1 = -(lambda + 0.2);
```

```julia
rho_1 = [rho_barrat_population(i, c, T, Np, ensemble, nsteps, epsilon, ep[1]) for i in l1];
```

```julia
for k in 2:length(ep)
    rho = [rho_barrat_population(i, c, T, Np, ensemble, nsteps, epsilon, ep[k]) for i in l1];
    rho_1 = hcat(rho_1,rho)
end
```

```julia
plot(lambda, rho_1[:,1], label = L"\epsilon ="*" $(ep[1])")
for i in 2:length(ep)
    plot(lambda, rho_1[:,i], label = L"\epsilon ="*"$(ep[i])")
end

plt[:xscale]("log")
plt[:yscale]("log")
#plt[:gca]()[:set_aspect]("equal")
plt[:grid](true)
#plt[:xticks](logspace(-1,-6,6))
#plt[:gca]()[:xaxis][:grid](true) 
plt[:legend]()
plt[:tight_layout]()
```

### Right side

```julia
l2 = (lambda - 0.2);
```

```julia
rho_2 = [rho_barrat_population(i, c, T, Np, ensemble, nsteps, epsilon, ep[1]) for i in l2];
```

```julia
for k in 2:length(ep)
    rho = [rho_barrat_population(i, c, T, Np, ensemble, nsteps, epsilon, ep[k]) for i in l2];
    rho_2 = hcat(rho_2,rho)
end
```

```julia
plot(lambda, rho_2[:,1], label = L"\epsilon ="*" $(ep[1])")
for i in 2:length(ep)
    plot(lambda, rho_2[:,i], label = L"\epsilon ="*"$(ep[i])")
end

plt[:xscale]("log")
plt[:yscale]("log")
#plt[:gca]()[:set_aspect]("equal")
plt[:grid](true)
#plt[:xticks](logspace(-1,-6,6))
#plt[:gca]()[:xaxis][:grid](true) 
plt[:legend]()
plt[:tight_layout]()
```

```julia
plot(l1, rho_1[:,1], label = L"\epsilon ="*" $(ep[1])"), yscale = :log10, xscale = :log10)
for i in 2:length(ep)
    plot!(-(lambda+0.2), rho_1[:,i], label = L"\epsilon ="*"$(ep[i])")
end
plot!()
```

```julia
plot(-(lambda+0.2)[3:end], rho_1[:,1][3:end], label = L"\epsilon ="*" $(ep[1])", yscale = :log10, xscale = :log10)
for i in 2:length(ep)
                plot!(-(lambda+0.2)[3:end], rho_1[:,i][3:end], label = L"\epsilon ="*"$(ep[i])")
end
plot!()
```

```julia
#Make the grid log in lambda-0.2
```

```julia
plot(ep,rho_1[1,:], yscale = :log10, xscale = :log10, label = "", xlabel = L"\epsilon",
ylabel = L"h(\lambda^*)")
plot!(ep,ep.^-1, label = L"\epsilon^{-1}")
plot!(ep,ep.^-0.9, label = L"\epsilon^{-0.9}")
```
