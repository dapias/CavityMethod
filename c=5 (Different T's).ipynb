```julia
Np = 2500
nsteps = 55000
epsilon = 1.e-300
ensemble = Int64(10^5);
```

```julia
epsilon2 = 1.e-3
```

```julia
c = 5;
```

```julia
T = [0.1,0.5,1.,2.,3.];
```

```julia
lambda_barrat = -linspace(-0.,1.01,200)
```

```julia
rho_ba = [rho_barrat_population(i, c, T[1], Np, ensemble, nsteps, epsilon, epsilon2) for i in lambda_barrat];
```

```julia
for k in 2:length(T)
    rho = [rho_barrat_population(i, c, T[k], Np, ensemble, nsteps, epsilon, epsilon2) for i in lambda_barrat];
    rho_ba = hcat(rho_ba,rho)
end
```

```julia
plot(lambda_barrat, rho_ba[:,1], label = L"T ="*"$(T[1])", yscale = :log10, xticks = -1.:0.2:0.)
for i in 2:length(T)
    plot!(lambda_barrat, rho_ba[:,i], label = L"T ="*"$(T[i])")
end
plot!()
```

```julia
plot(-lambda_barrat[2:end], rho_ba[:,1][2:end], label = L"T ="*"$(T[1])", yscale = :log10, xscale = :log10)
for i in 2:length(T)
    plot!(-lambda_barrat[2:end], rho_ba[:,i][2:end], label = L"T ="*"$(T[i])")
end
plot!(xlabel = L"-\lambda", ylabel = L"\rho(\lambda)")
```

```julia
plot(-lambda_barrat[2:end], -lambda_barrat[2:end].*rho_ba[:,1][2:end], label = L"T ="*"$(T[1])", yscale = :log10, xscale = :log10)
for i in 2:length(T)
    plot!(-lambda_barrat[2:end], -lambda_barrat[2:end].*rho_ba[:,i][2:end], label = L"T ="*"$(T[i])")
end
plot!(xlabel = L"-\lambda", ylabel = L"-lambda\rho(\lambda)")
```
