## Overview

```julia
T = 0.1
lambda_barrat = -logspace(-10.,-1, 100);
```

```julia
c = 5
```

```julia
epsilon2 = [1.e-3, 1.e-4,1.e-5,1.e-6, 1.e-10];
```

```julia
rho_ba = [rho_bouchaud_population(i, c, T, Np, ensemble, nsteps, epsilon, epsilon2[1]) for i in lambda_barrat];
```

```julia
for k in 2:length(epsilon2)
    rho = [rho_bouchaud_population(i, c, T, Np, ensemble, nsteps, epsilon, epsilon2[k]) for i in lambda_barrat];
    rho_ba = hcat(rho_ba,rho)
end
```

```julia
plot(-lambda_barrat, rho_ba[:,1], label = L"\epsilon ="*"$(epsilon2[1])", yscale = :log10, xscale = :log10);
```

```julia
for i in 2:length(epsilon2)
    plot!(-lambda_barrat, rho_ba[:,i], label = L"\epsilon ="*"$(epsilon2[i])")
end
plot!(-lambda_barrat,(-lambda_barrat).^(T-1)*T, label = L"(-\lambda)^{T-1}")
```

```julia
plot(epsilon2, rho_ba[1,:], xscale = :log10, yscale = :log10)
plot!(epsilon2, epsilon2.^-1., xscale = :log10, yscale = :log10)
plot!(epsilon2, epsilon2.^-0.86, xscale = :log10, yscale = :log10)
```

## First peak

```julia
#lambda = -linspace(0.19,0.21,100)
lambda = logspace(-6,-1,101);
```

```julia
ep = [1.e-3, 1.e-4,1e-5,1.e-6]
```

### Left side

```julia
l1 = -(lambda + 0.2)
```

```julia
rho_1 = [rho_barrat_population(i, c, T, Np, ensemble, nsteps, epsilon, ep[1]) for i in l1];
```

```julia
for k in 2:length(ep)
    rho = [rho_barrat_population(i, c, T, Np, ensemble, nsteps, epsilon, ep[k]) for i in l1];
    rho_1 = hcat(rho_1,rho)
end
```

```julia
plot(lambda, rho_1[:,1], label = L"\epsilon ="*" $(ep[1])", xscale = :log10, yscale = :log10)
for i in 2:length(ep)
    plot!(lambda, rho_1[:,i], label = L"\epsilon ="*"$(ep[i])")
end
plot!()
```

### Right side

```julia
l2 = (lambda - 0.2)
```

```julia
rho_2 = [rho_barrat_population(i, c, T, Np, ensemble, nsteps, epsilon, ep[1]) for i in l2];
```

```julia
for k in 2:length(ep)
    rho = [rho_barrat_population(i, c, T, Np, ensemble, nsteps, epsilon, ep[k]) for i in l2];
    rho_2 = hcat(rho_2,rho)
end
```

```julia
plot!(lambda, rho_2[:,1], label = L"\epsilon2 ="*" $(ep[1])", xscale = :log10, yscale = :log10)
for i in 2:length(ep)
    plot!(lambda, rho_2[:,i], label = L"\epsilon2 ="*"$(ep[i])")
end
plot!()
```

```julia
plot!()
```
